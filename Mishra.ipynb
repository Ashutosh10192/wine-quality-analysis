```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
red_wine_data = pd.read_csv('/content/winequality-red.csv', sep=';')
white_wine_data = pd.read_csv('/content/winequality-white.csv', sep=';')

# Remove duplicates
red_wine_data = red_wine_data.drop_duplicates()
white_wine_data = white_wine_data.drop_duplicates()

# Boxplots for outlier visualization
columns_to_plot = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol']
plt.figure(figsize=(15, 10))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 5, i)
    sns.boxplot(y=red_wine_data[col], color='red')
    plt.title(f'Red Wine - {col}')
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 5, i+5)
    sns.boxplot(y=white_wine_data[col], color='lightyellow')
    plt.title(f'White Wine - {col}')
plt.tight_layout()
plt.show()

# Quality distribution visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='quality', data=red_wine_data, color='red', edgecolor='black')
plt.title('Red Wine Quality Distribution')
plt.subplot(1, 2, 2)
sns.countplot(x='quality', data=white_wine_data, color='lightyellow', edgecolor='black')
plt.title('White Wine Quality Distribution')
plt.tight_layout()
plt.show()

# Correlation with quality visualization
corr_red = red_wine_data.corr()['quality'].drop('quality')
corr_white = white_wine_data.corr()['quality'].drop('quality')
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
corr_red.sort_values().plot(kind='bar', color='red')
plt.title('Correlation with Red Wine Quality')
plt.subplot(1, 2, 2)
corr_white.sort_values().plot(kind='bar', color='lightyellow')
plt.title('Correlation with White Wine Quality')
plt.tight_layout()
plt.show()
```
